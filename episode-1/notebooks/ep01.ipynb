```python
from lazypredict.Supervised import LazyClassifier, LazyRegressor
from sklearn.model_selection import train_test_split
from sklearn import datasets
import numpy as np
```

1. Regression Task

We will use Boston Housing data for this exercise

link: https://www.kaggle.com/code/prasadperera/the-boston-housing-dataset/notebook

```python
# load data
boston = datasets.load_boston()
X, y = boston.data, boston.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
print(f"X_train:{np.shape(X_train)}\tX_test: {np.shape(X_test)}\ty_train:{np.shape(y_train)}\ty_test: {np.shape(y_test)}")
```

```python
X_train
```

```python
y_train[0:20]
```

Build a Regression model

```python
# fit all models
reg = LazyRegressor(predictions=True)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)
```

```python
predictions
```

```python
y_test[0:10]
```

```python
pred_0 = predictions.iloc[0]
```

```python
pred_0
```

```python
predictions['actual'] = y_test
predictions.T
```

# Classification example

We will use the breast-cancer dataset for building the classification models

More info about the dataset: https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data

```python
# load data
data = datasets.load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
print(f"X_train:{np.shape(X_train)}\tX_test: {np.shape(X_test)}\ty_train:{np.shape(y_train)}\ty_test: {np.shape(y_test)}")
```

```python
y_test[0:10]
```

```python
# fit all models
clf = LazyClassifier(predictions=True)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
models
```

```python
predictions['actual'] = y_test
predictions.T
```

```python
pred_0 = predictions.iloc[0]
```

```python
pred_0.T
```
